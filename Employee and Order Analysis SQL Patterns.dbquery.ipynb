```sql
--Nth Highest Salary PASS WHERE 1,2,3 (WHICH EVER HIGHEST SALARY YOU WANT TO FETCH)
select distinct e1.salary
from employees e1
where 1 = (select count(distinct e2.salary)
from employees e2
where e2.salary >= e1.salary);

--Find Nth highest salary using dense rank
select salary
from (select salary, dense_rank() over (order by salary desc) as rnk
from employees
) where rnk = 3;

--Find the 2nd highest salary per department
select department, max(salary) as second_high_sal
from employees e
where salary < (select max(Salary) from employees where department = e.department
)
group by department;

--Find the 2nd highest salary per department
select * from(
  select employeeid, department, salary,
  dense_rank() over (partition by department order by salary desc) as rnk
  from employees
) where rnk = 2;

--Find employees who earn more than the average salary in their department
select employeeid, department, salary
from employees e
where salary > (select avg(salary) from employees where department = e.department);

--Get cumulative totalamount month-wise for each product
select productid, date_trunc('month', orderdate) as month,
sum(totalamount) over (PARTITION BY productid order by date_trunc('month', orderdate)
ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW) as cumulative_totalamount
from orders;

--Find customers who bought in January but not in February
SELECT DISTINCT customerid
FROM orders
WHERE MONTH(orderdate) = 1
EXCEPT
SELECT DISTINCT customerid
FROM orders
WHERE MONTH(orderdate) = 2;

--Detect duplicate records
SELECT customerid, COUNT(*) AS cnt
FROM customers
GROUP BY customerid
HAVING COUNT(*) > 1;

--Find top 3 selling products by totalamount
SELECT productid, SUM(totalamount) AS total_totalamount
FROM orders
GROUP BY productid
ORDER BY total_totalamount DESC
LIMIT 3;

--Customer rentention month over month

--Identify Inactive Customers
select o.customerid
from orders o
left join orders o1
on o.customerid = o1.customerid
group by o.CustomerID
having max(o1.orderdate) < current_date() - interval '3' month
or max(o1.orderdate) is null;

--Show top 3 highest selling products per category
select p.productid, p.category
from products p
join orders o
on o.ProductID = p.ProductID
group by p.productid, p.Category
order by sum(o.TotalAmount) desc
limit 3;

--Find duplicate email addresses in a customer table
SELECT email, COUNT(*) AS cnt
FROM customers
GROUP BY email
HAVING COUNT(*) > 1;

--Retrieve customers who placed orders in every month
select customerid
from orders
group by customerid
having count(distinct month(orderdate)) = 12;

--Retrieve customers who placed orders in every month of 2024
SELECT customerid
FROM orders
WHERE YEAR(orderdate) = 2025
GROUP BY customerid
HAVING COUNT(DISTINCT MONTH(orderdate)) = 12;

--Find products that have never been ordered
select p.productid, p.productname
from products p
left join orders o
on p.productid = o.productid
where o.productid is null;

--Find the top 3 customers with the highest total amount spent
select customerid
from customers c
join orders o
on c.customerid = o.customerid
group by customerid
order by sum(totalamount) desc
limit 3;
```
